```python
import pandas as pd


df_laptops = pd.read_csv('laptop_price.csv', encoding='latin1')
display(df_laptops.head())
```

```python
#1. Filtra las laptops que tienen un peso menor a 1.3 kg y además tienen un disco SSD de al menos 256GB.

# Limpiar la columna 'Weight' para obtener el valor numérico en kg
df_laptops['Weight_kg'] = df_laptops['Weight'].str.replace('kg', '').astype(float)

# Filtrar laptops con peso < 1.3 kg y SSD >= 256GB
mask = (
    
    (df_laptops['Weight_kg'] < 1.3) &
    (df_laptops['Memory'].str.contains('SSD', case=False, na=False)) &
    (df_laptops['Memory'].str.extract(r'(\d+)GB SSD', expand=False).fillna(0).astype(int) >= 256)
)

df_ssd_256_light = df_laptops[mask]
display(df_ssd_256_light[['Company', 'Product', 'Weight', 'Memory']])
```

---

### ¿Qué es una tabla pivote (o tabla dinámica)?

Una **tabla pivote** es una herramienta muy potente en hojas de cálculo (como Excel, Google Sheets, LibreOffice Calc, etc.) que sirve para **resumir, analizar, explorar y presentar grandes volúmenes de datos** de manera interactiva y flexible.

---

### Principales funciones y ventajas de una tabla pivote

- **Resumir datos:** Consolida grandes cantidades de datos en un formato más manejable y comprensible. Ejemplo: ver el total de ventas por mes, producto o vendedor.
- **Analizar datos:** Facilita la identificación de patrones, tendencias y anomalías. Permite comparar categorías y ver la contribución de cada elemento al total.
- **Explorar datos:** Permite "pivotar" los datos, es decir, cambiar la disposición para verlos desde diferentes perspectivas (filas, columnas, valores, filtros).
- **Agrupar datos:** Puedes agrupar fechas (por año, mes, trimestre), números (por rangos) o texto para un análisis más estructurado.
- **Calcular subtotales y totales generales:** Genera automáticamente estos cálculos.
- **Filtrar datos:** Permite aplicar filtros para enfocarse en subconjuntos específicos, como ventas de una región o período.
- **Presentar datos de forma clara:** Ideal para crear informes concisos, fáciles de entender y actualizables.

```python
import pandas as pd


df_laptops = pd.read_csv('laptop_price.csv', encoding='latin1')
display(df_laptops.head())
print(df_laptops.shape)
```

```python
"""1. Histograma del precio de las laptops
- Grafica un histograma de la columna `Price_euros` para ver la distribución de precios."""



plt.figure(figsize=(10,6))#tamaño de la figura
plt.hist(df_laptops['Price_euros'], bins=30, color='white', edgecolor='red')#caracteristicas del histograma
plt.title('Distribución de precios de laptops')
plt.xlabel('Precio en euros')
plt.ylabel('Cantidad de laptops')
plt.grid(axis='y', alpha=0.3)#significa que se muestra una cuadrícula (líneas de guía) solo en el eje Y del gráfico, con una transparencia del 30% (alpha=0.3)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.histplot(df_laptops['Price_euros'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribución de precios de laptops')
plt.xlabel('Precio en euros')
plt.ylabel('Cantidad de laptops')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
""" 2. Gráfico de barras: cantidad de laptops por compañía
- Muestra cuántos modelos tiene cada marca (`Company`) en un gráfico de barras."""

plt.figure(figsize=(10,6))
sns.countplot(data=df_laptops, x='Company', order=df_laptops['Company'].value_counts().index, palette='viridis')
plt.title('Cantidad de modelos por marca')
plt.xlabel('Marca (Company)')
plt.ylabel('Cantidad de modelos')
plt.xticks(rotation=45)
plt.show()
```

```python
"""3. Boxplot del precio por tipo de laptop
- Usa seaborn para graficar un boxplot de `Price_euros` según el tipo de laptop (`TypeName`)."""
plt.figure(figsize=(12,6))
sns.boxplot(data=df_laptops, x='TypeName', y='Price_euros', palette='Set2')
plt.title('Boxplot del precio por tipo de laptop')
plt.xlabel('Tipo de laptop (TypeName)')
plt.ylabel('Precio en euros')
plt.xticks(rotation=45)
plt.show()

```

```python
# 4. Gráfico de dispersión: precio vs. tamaño de pantalla

plt.figure(figsize=(10,6))
sns.scatterplot(data=df_laptops, x='Inches', y='Price_euros', hue='TypeName', palette='Set1', alpha=0.7)
plt.title('Precio vs. Tamaño de pantalla (Inches)')
plt.xlabel('Tamaño de pantalla (Inches)')
plt.ylabel('Precio en euros')
plt.legend(title='Tipo de laptop', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# 5. Gráfico de barras: promedio de RAM por compañía

df_laptops['Ram_GB'] = df_laptops['Ram'].str.replace('GB', '').astype(int)

ram_mean = df_laptops.groupby('Company')['Ram_GB'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=ram_mean.index, y=ram_mean.values, palette='mako')
plt.title('Promedio de RAM (GB) por compañía')
plt.xlabel('Compañía')
plt.ylabel('RAM promedio (GB)')
plt.xticks(rotation=45)
plt.show()
```

```python
# 6. Countplot de sistemas operativos


plt.figure(figsize=(10,6))
sns.countplot(data=df_laptops, x='OpSys', order=df_laptops['OpSys'].value_counts().index, palette='pastel')
plt.title('Cantidad de modelos por sistema operativo')
plt.xlabel('Sistema Operativo (OpSys)')
plt.ylabel('Cantidad de modelos')
plt.xticks(rotation=45)
plt.show()
```

```python
# 7. Boxplot del peso por tipo de laptop

# Limpiar la columna 'Weight' para obtener el valor numérico en kg
df_laptops['Weight_kg'] = df_laptops['Weight'].str.replace('kg', '').astype(float)


plt.figure(figsize=(12,6))
sns.boxplot(data=df_laptops, x='TypeName', y='Weight_kg', palette='Set3')
plt.title('Boxplot del peso por tipo de laptop')
plt.xlabel('Tipo de laptop (TypeName)')
plt.ylabel('Peso (kg)')
plt.xticks(rotation=45)
plt.show()
```

```python
# 8. Gráfico de barras apiladas: tipo de memoria por compañía
import matplotlib.pyplot as plt

memory_company = pd.crosstab(df_laptops['Company'], df_laptops['Memory'])
memory_company.plot(kind='bar', stacked=True, figsize=(12,7), colormap='tab20')
plt.title('Cantidad de modelos por tipo de memoria y compañía')
plt.xlabel('Compañía')
plt.ylabel('Cantidad de modelos')
plt.xticks(rotation=45)
plt.legend(title='Tipo de memoria', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

```

```python
# 9. Heatmap de correlación entre variables numéricas
import seaborn as sns

corr = df_laptops.select_dtypes(include=['float64', 'int64']).corr()
plt.figure(figsize=(10,7))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap de correlación entre variables numéricas')
plt.show() 
```

```python
# 10. Gráfico de violín: precio por cantidad de RAM
plt.figure(figsize=(10,6))
sns.violinplot(data=df_laptops, x='Ram_GB', y='Price_euros', palette='Set2')
plt.title('Precio de laptops según cantidad de RAM (GB)')
plt.xlabel('RAM (GB)')
plt.ylabel('Precio en euros')
plt.show()
```
